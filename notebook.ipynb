# Loading data

```python
#importing modules

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#dataframe
spotify = pd.read_csv('Spotify.csv')
```

```python
# most streamed artist

artist_by_streams = spotify[['Artist', 'Streams (Billions)']].head()
print(artist_by_streams)
```

```python
# most streams song 

song_by_treams = spotify[['Song', 'Streams (Billions)']].head()
print(song_by_treams)
```

```python
#the years with the most streams

year_by_streams = spotify[['Release Date', 'Streams (Billions)']].head()
print(year_by_streams)
```

```python
spotify['Artist'].value_counts(sort=True).head()
```

```python
spotify['Streams (Billions)'] = spotify['Streams (Billions)'].astype(str).str.replace(',', '').astype(float)
```

# Visualizing DataFrames

```python
#most streams artist
artist = spotify[['Artist', 'Streams (Billions)']].head(5)
print(artist)
```

```python
artist.plot(kind='bar',
            x='Artist', 
            y='Streams (Billions)', 
            title='Most Streams Artist')
plt.show()
```

```python
#Most Streams song

song = spotify[['Song', 'Streams (Billions)']].head(5)
print(song)
```

```python
song.plot(kind='barh',
          x='Song', 
          y='Streams (Billions)',
          title='Most Streams song' )
plt.show()
```

```python
#stream frequency

spotify['Streams (Billions)'].hist(bins=5)
plt.show()
```

```python
#most streamed release of the year
year = spotify.loc[:, ['Release Date', 'Streams (Billions)']].head(5)
print(year)
```

```python
year.plot(kind='line',
         x='Release Date',
         y='Streams (Billions)',
         title='Most streamed release of the year',
         rot=45)
plt.show()
```

```python
#comparing the two artists with the most streaming

artist_vs = spotify['Artist'].head(2)
print(artist_vs)
```

```python
spotify[spotify['Artist']=='Ed Sheeran']['Streams (Billions)'].hist(alpha=0.8)
spotify[spotify['Artist']=='The Weeknd']['Streams (Billions)'].hist(alpha=0.8)
plt.legend(['Ed Sheeran', 'The Weeknd'])
plt.show()
```
